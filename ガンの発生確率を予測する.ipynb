## ロジスティック回帰とは
ロジスティック関数（別名: シグモイド関数）とは､実数全体の値をとる入力を､0〜1の値に変換する関数です｡  
入力に線形入力をとって､出力の数値を確率とみなすことで利用できます｡  

<img src="images/logsitic.png" width="50%">
   
### オッズ
まずは､オッズについて解説します｡オッズ（オッズ比）とは､とあるできごとが起こる確率と起こらない確率の比です｡  
オッズのとる値の範囲は  0<odds<+∞  であり､確率のとる値の範囲を広げています｡  

<img src="images/ozz.png" width="50%">

### 対数オッズ
また､オッズの対数を取ったものを対数オッズ､あるいはロジット関数といいます｡  
対数の底は  e  (ネイピア数)であるとします｡対数オッズのとる値の範囲は  −∞<logit(odds)<+∞  であり､ 0<p<1  の範囲の値を取る確率を実数全体にまで広げることが出来ました｡  

<img src="images/no_ozz.png" width="50%">

### ロジスティック関数（シグモイド関数）
ロジスティック関数とは､ロジット関数の逆関数です｡ロジット関数とは逆に､実数全体を  0<p<1  の範囲に縮約できます｡  
このとき､入力である  logit  を線形入力置き換えて､2値分類における一方のクラスへの予測確率と捉えることで､予測確率を求めることが可能になります｡

<img src="images/log2.png" width="50%">

### 最尤推定
最尤推定とは､ロジスティック回帰の最適化の方法であり､最も尤も（もっとも）らしいパラメータを獲得するというコンセプトで名前がつけられています｡  
ロジスティック回帰も､線形分類器という面ではパーセプトロンと同じでしたが､この最適化手法の部分は大きく異なります｡   
最尤推定は誤差のとり方を｢正解したか､してないか｣という画一的な情報ではなく｢正解に対してどれだけ確信を持てたか｣という幅のある情報に変えることができ､パーセプトロンよりも柔軟に最適解に近づくことができます｡

# データの読み込み
今回のデータセットも､sklearn.datasetsに収録されているものを利用します｡  
Breast Cancer Wisconsin (Diagnostic) Data Setという名前の通り､1995年にアメリカの名門大学であるウィスコンシン大学で､乳がんの検査のために集められたデータセットです｡細胞核の画像を観察しれ得られた30の特徴量がまとめられています｡   
これらを説明変数として､確認された腫瘍が良性か悪性かを判定することができます｡また､今回はロジスティック回帰を利用しますので､発症しているかいないかの分類に加えて､発症の確率も出力します｡

```python
# 必要なライブラリの読み込み
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
import seaborn as sns
```

```python
# brest_cancerデータセットの読み込み
from sklearn.datasets import load_breast_cancer
bc = load_breast_cancer()
```

```python
# データの詳細を出力
print(bc.DESCR)
```

# データの観察

```python
# データをDataFrameに変換
df_data = pd.DataFrame(bc.data, columns=bc.feature_names)
df_target = pd.DataFrame(bc.target, columns=['class'])
df = pd.concat([df_data, df_target], axis=1)
df.head()
```

```python
# 基本統計量を出力
df.describe()
```

```python
# 散布図行列を出力（ここでは以下の４つの特徴量をピックアップ）
df_pickup = df.loc[:, ['worst perimeter', 'worst concave points', 'worst radius', 'mean concave points', 'class']]
sns.pairplot(df_pickup, size=2.0, hue='class', markers='+')
plt.show()
```

# 学習

```python
# 説明変数と目的変数を指定
X = df.loc[:, ['worst perimeter', 'mean concave points']].values
y = df.loc[:, ['class']].values
# scikit-learnの仕様に合わせて､一列のベクトルに変換
y = y.reshape(-1)
```

```python
# データの標準化（特徴量の値の大きさごとに重みが変わらないように標準化する）
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_std = scaler.transform(X)
```

```python
# データセットの分割
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.3, random_state=0)
```

```python
# ロジスティック回帰の実行
#ロジスティック回帰は､sklearn.linear_modelのLogisticRegressionクラスを利用して実行できます
#逆正則化パラメータ  C  は､デフォルトで1.0に設定されています｡

from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=1.0)
lr.fit(X_train, y_train)
```

# 学習結果の評価
今回は、 　　

・決定境界  

・正答率　　

を用います

```python
# ２次元データの決定曲線をプロットする関数
def plot_decision_regions(X, y, classifier, resolution=0.02):
    from matplotlib.colors import ListedColormap
    markers = ("s", "x", "o", "^", "v")
    colors = ("red", "blue", "lightgreen", "gray", "cyan")
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    x1_min, x1_max = X[:, 0].min()-1 , X[:, 0].max() +1
    x2_min, x2_max = X[:, 1].min()-1 , X[:, 1].max() +1
    
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution), np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
    
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1], alpha=0.8, c=cmap(idx),marker=markers[idx], label=cl)
```

```python
# 決定境界を出力(決定境界のプロットより､一応の分類性能があることが確認できます｡)
plot_decision_regions(X_std, y, lr)
plt.xlabel("petal length [standardizd]")
plt.ylabel("petal width [standardizd]")
plt.legend(loc = "upper left")
plt.show()
```

```python
# 正答率を出力
print('train acc: %.3f' % lr.score(X_train, y_train))
print('test acc: %.3f' % lr.score(X_test, y_test))
```

testの正答率は92.4%となりました｡また､trainの場合との比較から､過学習を起こしていないことが確認できます｡

# 学習モデルの利用
最尤推定により､ロジスティック回帰のモデルを正常に学習させることができました｡  
それでは､最後にこのモデルを実際の検査に利用することを想定して､testデータのサンプル（それぞれの患者の検査結果）に対して予測をさせてみましょう｡

```python
# testデータのサンプル数を出力
len(y_test)
```

```python
# testデータのインデックスを指定（0<=index<=170）
index = 0
```

```python
# 予測確率を出力
# [悪性である確率, 良性である確率]
lr.predict_proba(X_test[index].reshape(1, -1))
```
